```python
import pandas as pd
import re 
import math

day_4 = open('input.txt', 'r')
lines = day_4.readlines()
```

```python
x = re.compile('\w+-\w+-\w+')
```

```python
x.search('seed-to-soil').group()
```

```python
seeds = list(map(int,lines[0].split(':')[1].split(' ')[1:]))
```

```python
lines = [i.strip() for i in lines if i.strip() != '']
```

```python
line_order = [n for n, line in enumerate(lines) if x.match(line)]
```

```python
codes_length = [i - line_order[n-1]-1 for n, i in enumerate(line_order+[len(lines)]) if n > 0]
```

```python
codes_length_map = {lines[o]: l for o,l  in zip(line_order,codes_length)}
```

```python
codes_length_map.keys()
```

```python
dict_map = {line:[list(map(int, lines[i].split(' '))) for i in range(n+1,n+codes_length_map[line]+1,1)] for n, line in enumerate(lines) if line in codes_length_map.keys()}
```

```python
dict_map
```

```python
end_seeds = {k:None for k in seeds}
for s in seeds:
    start = s
    for k, v in dict_map.items():
        while v:
            code = v[0]
            if (start>=code[0]) & (start<code[0]+code[-1]):
                
                v = []
                start = code[1] + start - code[0]
                print(code, k,s,start)
            else:
                v=v[1:]

        # print(k,v,s)
        # start = start if v.get(start) == None else v.get(start)
        end_seeds[s] = start
```
