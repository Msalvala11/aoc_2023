```python
import pandas as pd
import re 

day_2 = open('input.txt', 'r')
lines = day_2.readlines()
```

only 12 red cubes, 13 green cubes, and 14 blue cubes

```python
def cube_assessment(colour, line):

    B = line.strip().split(':')[1]

    cubes = [int(i.split(' ')[0]) for i in re.findall(f'[0-9]+\s[{colour[0]}]', B)]
    
    greatest = cubes[0]
    while cubes:
        if cubes[0] > greatest:
            greatest = cubes[0]
        cubes = cubes[1:]
    
    print(greatest)
    return greatest
```

```python
def colour_loop(colour_list, line):

    squared = 1 

    while colour_list:
        squared = squared*cube_assessment(colour_list[0], line)
        colour_list = colour_list[1:]
        
    return squared
```

```python
colour_list = ['red', 'green', 'blue']
```

```python
lines[5]
```

```python
colour_loop(colour_list, lines[5])
```

```python
id_sum = 0
for line in lines:
    id_sum += colour_loop(colour_list, line)

print(id_sum)
```
