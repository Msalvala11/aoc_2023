```python
import pandas as pd
import re 

day_2 = open('input.txt', 'r')
lines = day_2.readlines()
```

only 12 red cubes, 13 green cubes, and 14 blue cubes

```python
def cube_assessment(colour, limit, line):

    A = line.strip().split(':')[0]
    B = line.strip().split(':')[1]

    cubes = [int(i.split(' ')[0]) for i in re.findall(f'[0-9]+\s[{colour[0]}]', B)]
    
    while cubes:
        if cubes[0] <= limit:
            cubes = cubes[1:]
        else:
            cubes = 0
            return 0
    
    return int(A.split(' ')[-1])

```

```python
def colour_loop(colour_dict, line):
    colour_list = list(colour_dict.keys())

    while colour_list:
        colour = colour_list[0]
        result = cube_assessment(colour, colour_dict[colour], line)
        
        if result != 0:
            colour_list = colour_list[1:]
        else:
            colour_list = 0
            return 0
        
    return result
```

```python
colour_dict = {
    'red':12,
    'green':13,
    'blue':14
}
```

```python
id_sum = 0
for line in lines:
    id_sum += colour_loop(colour_dict, line)

print(id_sum)
```
