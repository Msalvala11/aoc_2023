```python
import pandas as pd
import re 
import math

day_4 = open('input.txt', 'r')
lines = day_4.readlines()
```

```python
import re

x = re.compile('[0-9][0-9]|\s[0-9]\s|\s[0-9]$')
```

```python
points = 0
for line in lines:

    line = line.strip().split(':')[1]

    winners = [i.replace(' ','') for i in x.findall(line.split('|')[0])]
    if len(winners) != 10:
        print(winners, len(winners))
    card = [i.replace(' ','') for i in x.findall(line.split('|')[1])]
    if len(card) != 25:
        print(card, len(card))
    matches = 0
    while card:
        if card[0] in winners:
            matches += 1
        card = card[1:]

    points += math.floor(2**(matches-1))
    print(math.floor(2**(matches-1)), matches)
```
